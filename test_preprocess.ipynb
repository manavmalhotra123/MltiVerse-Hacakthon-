```python
test_data = pd.read_csv('test.csv')
```

```python
# Dropping the column that are not needed

import pandas as pd

# Assuming you have a DataFrame called 'df' containing the dataset

# List of columns to be removed
columns_to_remove = ['Kit Number','Unnamed: 0', 'ID', 'Name', 'Age', 'Photo', 'Nationality', 'Flag', 'Club', 'Club Logo', 'Wage', 'Special','Release Clause','DefensiveAwareness','Loaned From','Marking','Composure','Joined','Contract Valid Until','Work Rate','Real Face','Position','Jersey Number','Best Position','year']

# Drop the columns from the DataFrame
df = test_data.drop(columns=columns_to_remove)
```

```python
df.shape
```

```python
df['Body Type'] = df['Body Type'].str.split('(').str[0]

```

```python
# Perform label encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Body Type'] = le.fit_transform(df['Body Type'])
```

```python


df['Weight'] = df['Weight'].str.replace('lbs', '')
```

```python

df['Height'] = df['Height'].str.replace('cm', '')

```

```python

import pandas as pd
import numpy as np

# Assuming you have a dataset in a pandas DataFrame called 'df'
# and you want to fill the null values in the 'Volleys' column.

# Fill null values with mean
mean_value = df['Volleys'].mean()
df['Volleys'].fillna(mean_value, inplace=True)

# Alternatively, fill null values with median
median_value = df['Volleys'].median()
df['Volleys'].fillna(median_value, inplace=True)

```

```python
import numpy as np

# Assuming you have a dataset in a pandas DataFrame called 'df'
# and you want to fill the null values in the 'Volleys' column.

# Fill null values with mean
mean_value = df['Curve'].mean()
df['Curve'].fillna(mean_value, inplace=True)

# Alternatively, fill null values with median
median_value = df['Curve'].median()
df['Curve'].fillna(median_value, inplace=True)

```

```python
import pandas as pd
import numpy as np

# Assuming you have a dataset in a pandas DataFrame called 'df'
# and you want to fill the null values in the 'Curve' and 'Agility' columns.

# Fill null values in 'Curve' column with mean
mean_curve = df['Curve'].mean()
df['Curve'].fillna(mean_curve, inplace=True)

# Fill null values in 'Agility' column with average
average_agility = df['Agility'].mean()
df['Agility'].fillna(average_agility, inplace=True)

```

```python


# Find the most commonly occurring foot
most_common_foot = df['Preferred Foot'].mode().values[0]

# Fill missing values with the most commonly occurring foot
df['Preferred Foot'].fillna(most_common_foot, inplace=True)

df['Preferred Foot']

from sklearn.preprocessing import LabelEncoder

# Assuming you have a dataset in a pandas DataFrame called 'df'
# and you want to label encode the 'Preferred Foot' column.

# Perform label encoding
le = LabelEncoder()
df['Preferred Foot'] = le.fit_transform(df['Preferred Foot'])

```

```python
import pandas as pd

# Assuming you have a dataset in a pandas DataFrame called 'data'
# and you want to check the number of null values in each column.

# Check the number of null values in each column
null_counts = df.isnull().sum()/len(df)*100

# Save null value counts to a text file
null_counts.to_csv("test_null_percentage.txt", header=False)

print("Null value counts saved to 'null_percentage.txt'")
```
